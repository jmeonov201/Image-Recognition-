```python
matriz = np.array([[1,2,3],
                   [1,5,6],
                   [1,8,9],
                   [1,255,100],])
```

```python
matriz
```

```python
ecualizada = np.zeros_like(matriz)
```

```python
ecualizada
```

```python
alto = matriz.shape[0]
```

```python
alto 
```

```python
ancho = matriz.shape[1] 
```

```python
ancho
```

```python
h = [0.0]*256
```

```python
for i in range(alto):
    for j in range(ancho):
        h[matriz[i,j]] = h[matriz[i,j]]+1
```

```python
h
```

```python
h_norm = np.asarray(h) / (alto*ancho)
```

```python
h_norm
```

```python
h_cum = np.cumsum(h_norm)
```

```python
h_cum
```

```python
matriz[0,0]
```

```python
for i in range(alto):
    for j in range(ancho):
        pixel = matriz[i,j]    
        ecualizada[i,j] = 256 * h_cum[pixel]
```
