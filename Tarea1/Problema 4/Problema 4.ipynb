```python
import numpy as np ## para las matrices y vectores o tensores
```

```python
import seaborn as sb 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## creando la fucion
def main(imagen):
    imgStr = imagen ## creo un objeto y le asigno un valor string con el nombre de la imagen
    img = cv2.imread(imgStr) ## creo un nuevo objeto que va a leer la imagen creada en la fila de arriba  
    
    ## tamañao de la imagen
    alto = img.shape[0]
    ancho = img.shape[1]
    canales = img.shape[2]
    
    ## listas vacias donde se agregagaran las extracciones 
    blue = []
    green = []
    red = []
    grey = [] 
  
    ## llenado de gris la imagen
    for i in range(alto):
        for j in range(ancho):
            pixel = img[i,j] ## extraigo de la imagen en la posicion i, j, el pixel
            azul = pixel[0] ## del pixel saco el color azul
            verde = pixel[1] ## de pixel sacor el color verde
            rojo = pixel[2] ## de pixel sacor el color rojo
            grisaritmetico = int((int(pixel[0])+int(pixel[1])+int(pixel[2]))/3) ## calculo el pixel gris
        
            blue.append(azul)
            green.append(verde)
            red.append(rojo)
            grey.append(grisaritmetico)
            
    ## promedios
    blue_mean = np.mean(blue)
    green_mean = np.mean(green)
    red_mean = np.mean(red)
    grey_mean = np.mean(grey)
    
    ## graficas 
    ## lienzo
    fig  = plt.figure(figsize=(60,30))
    ax1 = fig.add_subplot(1,4,1)
    ax2 = fig.add_subplot(1,4,2)
    ax3 = fig.add_subplot(1,4,3)
    ax4 = fig.add_subplot(1,4,4)
    
    ## graficas
    ax1.hist(blue,color='b')
    ax1.axvline(blue_mean,linewidth=4, color='black')
    ax2.hist(green,color='g')
    ax2.axvline(green_mean,linewidth=4, color='black')
    ax3.hist(red,color='r')
    ax3.axvline(red_mean,linewidth=4, color='black')
    ax4.hist(grey,color='grey')
    ax4.axvline(grey_mean,linewidth=4, color='black')
    
    ## colocando titulos
    ax1.set_xlabel("Canal de color azul")
    ax1.set_title("Histograma-Canal de color azul")
    ax2.set_xlabel("Canal de color verde")
    ax2.set_title("Histograma-Canal de color verde")
    ax3.set_xlabel("Canal de color rojo")
    ax3.set_title("Histograma-Canal de color rojo")
    ax4.set_xlabel("Canal de color gris")
    ax4.set_title("Histograma-Canal de color gris")
    
    
    ## mostrando las graficas
    plt.show()
```

```python
main("noticia_normal.jpg")
```
